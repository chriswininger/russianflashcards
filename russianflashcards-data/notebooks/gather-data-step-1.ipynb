```python
import translators as ts

import csv

input_file_name = "./data/initial-words.csv"
output_file_name = "./data/translated-words.csv"

output="Word,Translation\n"
ndx = 0
# opening the file using "with"
# statement
with open(input_file_name, 'r') as data:
    for row in csv.DictReader(data):
        word = row['Word']

        translation = ts.server.bing(word, from_language='ru', to_language='en', professional_field='general')

        output = output + "{},{}\n".format(word, translation)

        ndx = ndx + 1
        print("processed word ({}): {} -> {}".format(ndx, word, translation))

with open(output_file_name, 'w') as file:
    file.write(output)
```

```python
# read back in progress and continure
import translators as ts
from IPython.display import display, clear_output
import csv
import time

input_file_name = "./data/initial-words.csv"
output_file_name = "./data/translated-words.csv"

output=""
ndx = 0

existing_words = None

with open(output_file_name, 'r') as file:
    existing_words = list(csv.DictReader(file))

def in_list(existing_words, word):
    matches = [x for x in existing_words if x['Word'] == word]
    
    return len(matches) > 0

with open(input_file_name, 'r') as data:
    for row in csv.DictReader(data):
        word = row['Word']

        if not in_list(existing_words, word):
            translation = ts.server.google(word, from_language='ru', to_language='en', professional_field='general')

            output = output + "{},{}\n".format(word, translation)

            ndx = ndx + 1
            
            clear_output(wait=True)
            print("processed word ({}): {} -> {}".format(ndx, word, translation))
            time.sleep(3)

print('done')
```
